```python
import pandas as pd
import matplotlib.pyplot as plt
import shap
import joblib
from lime.lime_tabular import LimeTabularExplainer
loaded = joblib.load('random_forest_and_linear_regression.joblib')
loaded.keys()
```

```python
X_train_df = loaded['X_train']
y_train = loaded['y_train']
X_test_df = loaded['X_test']
y_test = loaded['y_test']
preprocessor = loaded['preprocessor']
Random_Forest = loaded['RandomForest']
Linear_Regression = loaded['LinearRegresion']
```

```python
shap.initjs()
```

```python
X_shap = shap.sample(X_train_df, nsamples = 400, random_state = 321)
X_shap.head()
```

```python
y_shap = y_train.loc[X_shap.index]
y_shap
```

```python
explainer = shap.Explainer(Random_Forest)
explainer
```

```python
shap_values = explainer(X_shap, y_shap)
type(shap_values)
```

```python
shap.summary_plot(shap_values, X_shap)
```

## Local Explanations

```python
row = 99
shap.force_plot(explainer.expected_value,
               shap_values = shap_values[row].values,
               features = X_shap.iloc[row])
```

```python
lime_explainer = LimeTabularExplainer(training_data = X_shap.values,
                                     feature_names = X_shap.columns, mode = 'regression')

lime_explainer
```

```python
exp = lime_explainer.explain_instance(X_shap.iloc[row],
                                     Random_Forest.predict)
exp.show_in_notebook()
```
